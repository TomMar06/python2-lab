```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Seed zapewnia identyczne dane u każdego studenta
np.random.seed(42)
n = 200

dzialy = np.random.choice(['IT', 'Sprzedaz', 'HR', 'Marketing', 'Finanse'], n, p=[0.30, 0.25, 0.15, 0.20, 0.10])
staz = np.random.gamma(shape=3, scale=2, size=n).clip(0.5, 20).round(1)
baza = {'IT': 9000, 'Sprzedaz': 7000, 'HR': 6500, 'Marketing': 7500, 'Finanse': 8500}

wynagrodzenie = np.array([
    baza[d] + staz[i] * 300 + np.random.normal(0, 1200)
    for i, d in enumerate(dzialy)
]).clip(4000, 25000).round(-2)

# 5 celowo wstawionych outlierów (błędy danych / kontrakty specjalne)
wynagrodzenie[np.random.choice(n, 5, replace=False)] = np.random.choice(
    [2000, 2500, 35000, 40000, 38000], 5, replace=False
)

# Tworzenie DataFrame
df = pd.DataFrame({
    'dzial': dzialy,
    'staz_lat': staz,
    'wynagrodzenie': wynagrodzenie,
    'wiek': (25 + staz + np.random.normal(0, 3, n)).clip(22, 65).round().astype(int),
    'ocena_roczna': np.random.choice([1, 2, 3, 4, 5], n, p=[0.05, 0.10, 0.40, 0.35, 0.10])
})

print(f"Dataset HR: {df.shape[0]} pracowników, {df.shape[1]} kolumn")
print(df.head())
print("\nTypy kolumn:")
print(df.dtypes)
```

```python
# 1a. Miary tendencji centralnej
placa = df['wynagrodzenie']
srednia = placa.mean()
mediana = placa.median()
dominanta = placa.mode().iloc[0]

print("=== MIARY TENDENCJI CENTRALNEJ ===")
print(f"Średnia:   {srednia:>10,.0f} PLN")
print(f"Mediana:   {mediana:>10,.0f} PLN")
print(f"Dominanta: {dominanta:>10,.0f} PLN")

# ODPOWIEDŹ NA PYTANIE: Czy mediana < średnia? Co to mówi o rozkładzie?
# Mediana jest niższa od średniej, ponieważ rozkład jest prawostronnie skośny. 
# Kilku pracowników z bardzo wysokimi zarobkami (wprowadzone anomalie / kontrakty specjalne) 
# mocno ciągnie średnią w górę, natomiast mediana (jako miara pozycyjna) pozostaje na nie odporna.

# 1b. Miary rozproszenia
odch_std = placa.std()
q1 = placa.quantile(0.25)
q3 = placa.quantile(0.75)
iqr = q3 - q1
rozstep = placa.max() - placa.min()

print("\n=== MIARY ROZPROSZENIA ===")
print(f"Odchylenie std: {odch_std:>10,.0f} PLN")
print(f"Q1 (P25):       {q1:>10,.0f} PLN")
print(f"Q3 (P75):       {q3:>10,.0f} PLN")
print(f"IQR:            {iqr:>10,.0f} PLN")
print(f"Rozstęp:        {rozstep:>10,.0f} PLN")

# ODPOWIEDŹ NA PYTANIE: Dlaczego rozstęp jest tak duży?
# Rozstęp jest ogromny (38 000 PLN), ponieważ uwzględnia skrajne anomalie w danych — 
# najniższą pensję 2 000 PLN oraz najwyższą 40 000 PLN.

# 1c. Statystyki per dział
dzialy_stats = df.groupby('dzial', observed=True)['wynagrodzenie'].agg([
    'mean', 'median', 'std'
]).round().sort_values('median', ascending=False)
dzialy_stats.columns = ['Średnia', 'Mediana', 'Std']

print("\n=== WYNAGRODZENIA PER DZIAŁ ===")
print(dzialy_stats)

# 1d. Wizualizacja - histogram z miarami centralnymi
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(df['wynagrodzenie'], bins=20, color='steelblue', alpha=0.8, edgecolor='white')

ax.axvline(srednia, color='red', linestyle='--', lw=2, label=f'Średnia: {srednia:,.0f} PLN')
ax.axvline(mediana, color='green', linestyle='-', lw=2, label=f'Mediana: {mediana:,.0f} PLN')
ax.axvline(dominanta, color='orange', linestyle=':', lw=2, label=f'Dominanta: {dominanta:,.0f} PLN')

ax.set_title('Rozkład wynagrodzeń — miary tendencji centralnej')
ax.set_xlabel('Wynagrodzenie (PLN)')
ax.set_ylabel('Liczba pracowników')
ax.legend()
plt.tight_layout()
plt.show()
plt.close()
```

```python
# 2a. Korelacja Pearsona: staż vs wynagrodzenie
r, p_value = stats.pearsonr(df['staz_lat'], df['wynagrodzenie'])
print("=== KORELACJA PEARSONA ===")
print(f" r = {r:.4f}")
print(f" p = {p_value:.4f}")
print(f" Interpretacja: {'istotna' if p_value < 0.05 else 'nieistotna'} statystycznie")

# 2b. Korelacja Spearmana - porównanie
rho, p_rho = stats.spearmanr(df['staz_lat'], df['wynagrodzenie'])
print("\n=== KORELACJA SPEARMANA ===")
print(f" rho = {rho:.4f}")
print(f" p = {p_rho:.4f}")
print(f"\nPorównanie: Pearson r={r:.3f} vs Spearman rho={rho:.3f}")
print(f"Różnica: {abs(rho - r):.3f} -> {'duże' if abs(rho - r) > 0.1 else 'małe'} rozbieżności")

# ODPOWIEDŹ NA PYTANIE: Dlaczego Spearman jest wyższy niż Pearson?
# Spearman daje wyższy wynik, ponieważ bazuje na rangach obserwacji, dzięki czemu 
# jest odporny na wartości odstające (outliery), które silnie zaniżają liniowy współczynnik Pearsona.

# 2c. Macierz korelacji - wszystkie zmienne
corr = df[['staz_lat', 'wynagrodzenie', 'wiek', 'ocena_roczna']].corr()
print("\nMacierz korelacji Pearsona:")
print(corr.round(3))

# Wizualizacja ręczna (matplotlib bez seaborn)
fig, ax = plt.subplots(figsize=(7, 6))
im = ax.imshow(corr.values, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(im, ax=ax)

ax.set_xticks(range(len(corr.columns)))
ax.set_yticks(range(len(corr.columns)))
ax.set_xticklabels(corr.columns, rotation=45, ha='right')
ax.set_yticklabels(corr.columns)

# Dodanie wartości liczbowych do komórek
for i in range(len(corr)):
    for j in range(len(corr.columns)):
        ax.text(j, i, f'{corr.values[i, j]:.2f}',
                ha='center', va='center', fontsize=11,
                color='black' if abs(corr.values[i, j]) < 0.7 else 'white')

ax.set_title('Macierz korelacji — dataset HR')
plt.tight_layout()
plt.show()
plt.close()

# 2d. Korelacja: marketing spend vs revenue
np.random.seed(100)
n_mkt = 50
marketing_spend = np.random.uniform(10000, 100000, n_mkt)
revenue = marketing_spend * 5.2 + np.random.normal(0, 30000, n_mkt)
revenue = revenue.clip(0)

# Oblicz korelację Pearsona
r_mkt, p_mkt = stats.pearsonr(marketing_spend, revenue)

# Wykres rozrzutu z linią trendu
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(marketing_spend, revenue, alpha=0.6, color='coral', s=50)

# Linia trendu (polyfit stopień 1)
z = np.polyfit(marketing_spend, revenue, 1)
p_fit = np.poly1d(z)
x_line = np.linspace(marketing_spend.min(), marketing_spend.max(), 100)
ax.plot(x_line, p_fit(x_line), 'b', lw=2, label=f'Trend (r={r_mkt:.2f})')

ax.set_title('Wydatki na marketing vs Przychody')
ax.set_xlabel('Wydatki na marketing (PLN)')
ax.set_ylabel('Przychody (PLN)')
ax.legend()
plt.tight_layout()
plt.show()
plt.close()

print(f"Marketing spend vs Revenue: r = {r_mkt:.4f}, p = {p_mkt:.2e}")
```

```python
# 3a. scipy.stats.describe - szybki przegląd rozkładu
opis = stats.describe(df['wynagrodzenie'])
print("=== SCIPY STATS DESCRIBE ===")
print(f"Liczba obserwacji (nobs): {opis.nobs}")
print(f"Minimum:                  {opis.minmax[0]:,.0f} PLN")
print(f"Maximum:                  {opis.minmax[1]:,.0f} PLN")
print(f"Średnia:                  {opis.mean:,.0f} PLN")
print(f"Wariancja:                {opis.variance:,.0f}")
print(f"Skośność (skewness):      {opis.skewness:.3f}")
print(f"Kurtoza (kurtosis):       {opis.kurtosis:.3f}")

# INTERPRETACJA SŁOWNA:
# Skośność wynosi ok. 5.10 (> 1.0), co oznacza rozkład silnie prawoskośny (długi prawy ogon).
# Kurtoza wynosi ok. 34.16 (> 0), co oznacza rozkład wysoce spiczasty o "grubych ogonach", 
# potwierdzający obecność nagłych i skrajnych wartości odstających w zbiorze danych.

# 3b. Analiza percentylowa
procenty = [10, 25, 50, 75, 90, 95, 99]
percentyle = np.percentile(df['wynagrodzenie'], procenty)

print("\n=== ANALIZA PERCENTYLOWA ===")
for p, perm in zip(procenty, percentyle):
    print(f"P{p}: {perm:>10,.0f} PLN")

# ODPOWIEDZI NA PYTANIA BIZNESOWE:
p75_rynkowy = df['wynagrodzenie'].quantile(0.75)
procent_powyzej_12k = (df['wynagrodzenie'] > 12000).mean() * 100

print(f"\nBenchmark P75 wynosi: {p75_rynkowy:,.0f} PLN")
print(f"Powyżej 12 000 PLN zarabia: {procent_powyzej_12k:.1f}% pracowników.")

# 3c. Analiza per dział - pełna tabela
dzialy_pelna = df.groupby('dzial', observed=True)['wynagrodzenie'].agg([
    'mean', 'median', 'std', 'min', 'max', lambda x: x.quantile(0.75) - x.quantile(0.25)
]).round().sort_values('median', ascending=False)
dzialy_pelna.columns = ['Średnia', 'Mediana', 'Std', 'Min', 'Max', 'IQR']

print("\n=== PEŁNA TABELA STATYSTYK PER DZIAŁ ===")
print(dzialy_pelna)

# 3d. Wizualizacja - boxplot per dział (wbudowany pandas, bez seaborn)
fig, ax = plt.subplots(figsize=(10, 6))
df.boxplot(column='wynagrodzenie', by='dzial', ax=ax)
ax.axhline(df['wynagrodzenie'].median(), color='red', linestyle='--', label='Mediana globalna')

ax.set_title('Rozkład wynagrodzeń per dział')
plt.suptitle('')  # Usunięcie automatycznego, nadmiarowego tytułu pandas
ax.set_xlabel('Dział')
ax.set_ylabel('Wynagrodzenie (PLN)')
ax.legend()
plt.tight_layout()
plt.show()
plt.close()

# 3e. Opis rozkładu per dział IT i HR
print("\n=== SKOŚNOŚĆ I KURTOZA: IT vs HR ===")
for d in ['IT', 'HR']:
    podzb = df[df['dzial'] == d]['wynagrodzenie']
    opis_d = stats.describe(podzb)
    print(f"Dział {d} (n={opis_d.nobs}):")
    print(f"  Skośność: {opis_d.skewness:.3f}")
    print(f"  Kurtoza:  {opis_d.kurtosis:.3f}")

# INTERPRETACJA WYNIKU:
# Dział IT ma bardzo silną skośność dodatnią (~5.07) oraz ogromną kurtozę przez ulokowane tam skrajne outliery.
# Dział HR ma skośność bliską zera (~ -0.46) i lekką ujemną kurtozę, co oznacza rozkład w miarę symetryczny, stabilny i pozbawiony anomalii płacowych.
```

```python
# 4a. Wykrywanie outlierów metodą IQR
q1 = df['wynagrodzenie'].quantile(0.25)
q3 = df['wynagrodzenie'].quantile(0.75)
iqr = q3 - q1
dolna = q1 - 1.5 * iqr
gorna = q3 + 1.5 * iqr

print("=== METODA IQR (REGUŁA TUKEYA) ===")
print(f"IQR granice: [{dolna:,.0f} PLN, {gorna:,.0f} PLN]")

maska_iqr = (df['wynagrodzenie'] < dolna) | (df['wynagrodzenie'] > gorna)
outliery_iqr = df[maska_iqr]
print(f"Liczba outlierów IQR: {maska_iqr.sum()} z {len(df)}")
print("\nSzczegóły outlierów IQR:")
print(outliery_iqr[['dzial', 'staz_lat', 'wynagrodzenie']].to_string())

# 4b. Wykrywanie outlierów metodą z-score
z_scores = np.abs(stats.zscore(df['wynagrodzenie']))
maska_z = z_scores > 3.0
outliery_z = df[maska_z]

print("\n=== METODA Z-SCORE (|z| > 3.0) ===")
print(f"Outlierzy z-score: {maska_z.sum()} obserwacji")
print(outliery_z[['dzial', 'staz_lat', 'wynagrodzenie']].to_string())

# 4c. Wpływ outlierów na statystyki
bez_outlierow = df[~maska_iqr]['wynagrodzenie']
z_outlierami = df['wynagrodzenie']

print(f"\n{'Miara': <20} {'Z outlierami':>15} {'Bez outlierów':>15} {'Zmiana':>10}")
print("-" * 62)

for nazwa, f_z, f_bez in [
    ('Średnia', z_outlierami.mean(), bez_outlierow.mean()),
    ('Mediana', z_outlierami.median(), bez_outlierow.median()),
    ('Std', z_outlierami.std(), bez_outlierow.std())
]:
    zmiana = f_z - f_bez
    print(f"{nazwa: <20} {f_z:>15,.0f} {f_bez:>15,.0f} {zmiana:>+10,.0f}")

# ODPOWIEDŹ NA PYTANIE: Która miara jest bardziej stabilna? Dlaczego?
# Mediana okazała się najbardziej stabilną miarą (odporną) — jej różnica po usunięciu outlierów wynosi zaledwie 50 PLN.
# Odchylenie standardowe (std) reaguje dramatycznie — po usunięciu wartości odstających zmalało ponad dwukrotnie (o 2182 PLN).
# Wynika to z faktu, że std podnosi odchylenia do kwadratu, dając ogromną wagę obserwacjom skrajnym.
```
