## Ćwiczenie 3


```python
nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka']
ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

Zadanie 1

```python
# Średnia cena, najdroższy produkt, najtańszy produkt
print(f"Średnia cena to: {ceny.mean()}"),
print(f"Najwyższa cena to: {ceny.max()}")    ,
print(f"Najniższa cena to: {ceny.min()}"),
print(f"{ceny.argmax()}"), 
print(f"{ceny.argmin()}")
```

Zadanie 2

```python
# Obliczanie wartości dla każdego produktu
wartosc_produktow = ceny * ilosci_magazyn

# Łączna wartość magazynu
laczna_wartosc = wartosc_produktow.sum()

print(f"Wartości poszczególnych pozycji: {wartosc_produktow}")
print(f"Łączna wartość magazynu: {laczna_wartosc:.2f} zł")
```

Zadanie 3

```python
cena_rabat = ceny * 0.85
print(f"Ceny po rabacie: {cena_rabat}")
cena_VAT = ceny * 1.23
print(f"Ceny z VAT: {cena_VAT}")
# Ile produktów kosztuje powyżej 1000 zł?
print(f"Ceny powyżej 1000: {(ceny > 1000).sum()}")

```

Zadanie 4


```python
# KROK 0: Konwersja listy nazw na obiekt numpy array, aby móc używać masek
nazwy_arr = np.array(nazwy)

maska_oceny = oceny_klientow >= 4.5
produkty_premium_nazwy = nazwy_arr[maska_oceny]
produkty_premium_ceny = ceny[maska_oceny]

maska_tanie = ceny < 200
maska_tanie_nazwy = nazwy_arr[maska_tanie]
maska_tanie_ilość = ilosci_magazyn[maska_tanie]

print("Produkty z wysoką oceną (>= 4.5):")
for n, c in zip(produkty_premium_nazwy, produkty_premium_ceny):
    print(f"- {n}: {c} zł")
    
print("Produkty tańsze niż 200zł:")
for n, i in zip(maska_tanie_nazwy, maska_tanie_ilość):
    print(f"- {n}: {i}")
```

Zadanie 5

```python

nazwy_arr = np.array(nazwy)

wartosc_kazdego = ceny * ilosci_magazyn 
laczna_wartosc = wartosc_kazdego.sum()

maska_niski_stan = ilosci_magazyn < 30
produkty_braki = nazwy_arr[maska_niski_stan]

wartosc_brakow = wartosc_kazdego[maska_niski_stan].sum()
procent_wartosci = (wartosc_brakow / laczna_wartosc) * 100

print(f"Produkty wymagające zamówienia (< 30 szt.): {', '.join(produkty_braki)}")
print(f"Wartość tych produktów: {wartosc_brakow:,.2f} zł")
print(f"Stanowią one {procent_wartosci:.2f}% wartości całego magazynu.")
```

```python
maska_okazja = (oceny_klientow >= 4.5) & (ceny < 1000)
okazje_nazwy = nazwy_arr[maska_okazja]
okazje_ceny = ceny[maska_okazja]

maska_wysoka_ocena = oceny_klientow >= 4.0
maska_niska_ocena = oceny_klientow < 4.0

srednia_wysokie = ceny[maska_wysoka_ocena].mean()
srednia_niskie = ceny[maska_niska_ocena].mean()

print("Produkty będące okazją (ocena >= 4.5 i cena < 1,000 zł):")
for n, c in zip(okazje_nazwy, okazje_ceny):
    print(f"- {n}: {c} zł")

print(f"\nŚrednia cena produktów z wysoką oceną (>= 4.0): {srednia_wysokie:.2f} zł")
print(f"Średnia cena produktów z niską oceną (< 4.0): {srednia_niskie:.2f} zł")
```

## Ćwiczenie 4

```python
import time

# Przygotowanie danych
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)
```

```python
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter() - start
start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter() - start
print(f"=== Mnożenie ×2 ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
# Test 2: Suma
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter() - start
start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter() - start
print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
# Test 3: Filtrowanie (elementy > 500000)
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter() - start
start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter() - start
print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```
