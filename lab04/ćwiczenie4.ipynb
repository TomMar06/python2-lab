```python
# Ćwiczenie 3
np.random.seed(2026)
# 8 pracowników, dane roczne
pracownicy = ['Anna', 'Jan', 'Ewa', 'Marek', 'Kasia', 'Piotr', 'Zofia', 'Tomek']
pensje = np.array([5500, 7200, 4800, 9100, 6300, 4200, 8500, 5800])
staz = np.array([3, 8, 2, 15, 5, 1, 12, 4])
oceny = np.array([4.2, 3.8, 4.5, 4.0, 4.7, 3.5, 4.3, 3.9])
# Sprzedaż miesięczna 4 produktów przez 6 miesięcy
produkty_nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Akcesorium']
sprzedaz = np.random.randint(20, 200, size=(4, 6))
ceny_produktow = np.array([3500, 1800, 2500, 150])
```

```python
# Zadanie 3a
# 1. Oblicz średnią, medianę i odchylenie standardowe pensji
print (f"Średnia: {pensje.mean()}") 
print (f"Mediana: {np.median(pensje)}")
print (f"Odchylenie: {pensje.std()}")
# Podpowiedź: pensje.mean(), np.median(pensje), pensje.std()
# 2. Wyświetl imiona osób zarabiających powyżej średniej
nazwiska = np.array(pracownicy)
print(nazwiska[pensje > pensje.mean()])

# Uwaga: pracownicy to lista — zamień na np.array(pracownicy), żeby użyć maski
# Podpowiedź: nazwy = np.array(pracownicy), potem nazwy[pensje > pensje.mean()]
# 3. Wyświetl imiona osób zarabiających poniżej mediany

nazwiska = np.array(pracownicy)

mediana = np.median(pensje)

print(nazwiska[pensje < mediana])
# 4. Oblicz rozstęp (max - min) i rozstęp międzykwartylowy (IQR)
# Podpowiedź: IQR = np.percentile(pensje, 75) - np.percentile(pensje, 25)
# Rozstęp (cały zakres: od najniższej do najwyższej pensji)
rozstep = pensje.max() - pensje.min()
iqr = np.percentile(pensje, 75) - np.percentile(pensje, 25)
print(f"Rozstęp: {rozstep}")
print(f"IQR: {iqr}")
```

```python
# Zadanie 3b
# 1. Obliczanie premii na podstawie ocen
premie = np.where(oceny >= 4.5, 2000, 
                  np.where(oceny >= 4.0, 1000, 0))

# 2. Obliczanie nowego wynagrodzenia (pensja + premia)
nowe_wynagrodzenia = pensje + premie

# 3. Wyświetlenie wyników dla wszystkich pracowników
print("Premie poszczególnych osób:", premie)
print("Nowe wynagrodzenia:", nowe_wynagrodzenia)

# 4. Kto dostał najwyższą premię?
# np.argmax(premie) zwraca indeks pierwszej osoby z najwyższą kwotą
indeks_najwyzszej_premii = np.argmax(premie)
pracownik_z_premia = np.array(pracownicy)[indeks_najwyzszej_premii]

print(f"\nNajwyższą premię ({premie.max()} zł) otrzymał/a: {pracownik_z_premia}")
```

```python
# Zadanie 3c

kolejnosc = np.argsort(pensje)[::-1]

print(f"{'Poz.':<5} | {'Imię':<10} | {'Pensja':<8} | {'Staż':<5}")
print("-" * 35)

for i, idx in enumerate(kolejnosc, 1):
    imie = pracownicy[idx]
    pensja = pensje[idx]
    lata_stazu = staz[idx]
    
    print(f"{i:<5} | {imie:<10} | {pensja:<8} | {lata_stazu:<5}")
```

```python
# Zadanie 3d
kor_staz_pensja = np.corrcoef(staz, pensje)[0, 1]

kor_staz_ocena = np.corrcoef(staz, oceny)[0, 1]

print(f"Korelacja Staż vs Pensja: {kor_staz_pensja:.4f}")
print(f"Korelacja Staż vs Ocena: {kor_staz_ocena:.4f}")


if abs(kor_staz_pensja) > abs(kor_staz_ocena):
    print("\nSilniejszy związek występuje między stażem a pensją.")
else:
    print("\nSilniejszy związek występuje między stażem a oceną.")
```

```python
# Zadanie 3e
# Zmiana kształtu cen na (4, 1)
ceny_kol = ceny_produktow.reshape(4, 1)

# Mnożenie macierzy przez kolumnę (Broadcasting)
przychod = sprzedaz * ceny_kol

print("Macierz przychodu (produkty x miesiące):")
print(przychod)
# Łączny przychód na produkt
suma_produkty = przychod.sum(axis=1)
najlepszy_produkt_idx = np.argmax(suma_produkty)
najlepszy_produkt_nazwa = produkty_nazwy[najlepszy_produkt_idx]

# Łączny przychód na miesiąc
suma_miesiace = przychod.sum(axis=0)
najlepszy_miesiac_idx = np.argmax(suma_miesiace)

print(f"\nNajlepszy produkt: {najlepszy_produkt_nazwa} ({suma_produkty.max()} zł)")
print(f"Najlepszy miesiąc (indeks): {najlepszy_miesiac_idx} (Suma: {suma_miesiace.max()} zł)")
# Podział na kwartały/półrocza (pierwsze 3 vs ostatnie 3 kolumny)
pierwsza_polowa = przychod[:, :3].sum()
druga_polowa = przychod[:, 3:].sum()

print(f"\nPrzychód - pierwsze 3 miesiące: {pierwsza_polowa} zł")
print(f"Przychód - ostatnie 3 miesiące: {druga_polowa} zł")

if druga_polowa > pierwsza_polowa:
    roznica = ((druga_polowa / pierwsza_polowa) - 1) * 100
    print(f"Trend: Wzrostowy! Sprzedaż wzrosła o {roznica:.2f}%")
else:
    print("Trend: Spadkowy.")
```

```python
# Zadanie 4
import numpy as np

# --- Krok 1 — Generowanie ---
np.random.seed(42)

# Wiek: rozkład normalny, średnia 35, std 10, min 18
wiek = np.clip(np.random.normal(35, 10, 100).astype(int), 18, None)

# Wydatki: rozkład normalny, średnia 500, std 200, min 0
wydatki = np.clip(np.random.normal(500, 200, 100), 0, None)

# Liczba wizyt: rozkład Poissona, lambda=5
wizyty = np.random.poisson(5, 100)

# --- Krok 2 — Analiza ---

### 1. Statystyki opisowe
def drukuj_statystyki(nazwa, dane):
    print(f"{nazwa:10} | Średnia: {dane.mean():.1f} | Mediana: {np.median(dane):.1f} | "
          f"Std: {dane.std():.1f} | Min: {dane.min():.1f} | Max: {dane.max():.1f}")

print("--- Statystyki opisowe ---")
drukuj_statystyki("Wiek", wiek)
drukuj_statystyki("Wydatki", wydatki)
drukuj_statystyki("Wizyty", wizyty)
print("-" * 30)

### 2 i 3. Korelacje
kor_wiek_wydatki = np.corrcoef(wiek, wydatki)[0, 1]
kor_wizyty_wydatki = np.corrcoef(wizyty, wydatki)[0, 1]

print(f"Korelacja Wiek vs Wydatki: {kor_wiek_wydatki:.3f}")
print(f"Korelacja Wizyty vs Wydatki: {kor_wizyty_wydatki:.3f}")
print("-" * 30)

### 4. Segmentacja klientów
sr = wydatki.mean()
sd = wydatki.std()

# Zagnieżdżony np.where
segmenty = np.where(wydatki > sr + sd, 'WYSOKI', 
           np.where(wydatki < sr - sd, 'NISKI', 'ŚREDNI'))

### 5. Liczba klientów w segmentach
nazwy_segmentow, liczebnosc = np.unique(segmenty, return_counts=True)

print("Podział na segmenty:")
for nazwa, ilosc in zip(nazwy_segmentow, liczebnosc):
    print(f" - {nazwa}: {ilosc} klientów")
```
