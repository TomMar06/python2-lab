```python
import pandas as pd
import numpy as np

oceny = pd.Series([4.5, 3.0, 5.0, 3.5, 4.0])

dane_pensje = {
    'Anna': 5500,
    'Jan': 7200,
    'Ewa': 4800,
    'Marek': 9100,
    'Kasia': 6300
}
pensje = pd.Series(dane_pensje)

print("Series Pensje:")
print(pensje)
```

```python
najwiecej_imie = pensje.idxmax()
print(f"Najwięcej zarabia: {najwiecej_imie}")

srednia_pensja = pensje.mean()
print(f"Średnia pensja: {srednia_pensja:.1f}")

powyzej_sredniej = pensje[pensje > srednia_pensja]
print("Osoby powyżej średniej:")
print(powyzej_sredniej.index.tolist())

ponizej_6000 = (pensje < 6000).sum()
print(f"Liczba osób zarabiających < 6000: {ponizej_6000}")
```

```python
podwyzka_10 = pensje * 1.10

premia_500 = pensje + 500

nowe_pensje = (pensje * 1.10) + 500

ranking = nowe_pensje.sort_values(ascending=False)

print("\nRanking nowych pensji (od najwyższej):")
print(ranking)

```

```python
# Zadanie 3
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
penguins = pd.read_csv(url)
# Zadanie 3a: Pierwszy rzut oka
print("--- Zadanie 3a ---")
display(penguins.head(10))
print(f"Shape: {penguins.shape}")
penguins.info() 
# Info pokaże braki: np. 'sex' ma 333 wartości non-null na 344 wiersze (11 braków).

# Zadanie 3b: Statystyki opisowe
print("\n--- Zadanie 3b ---")
display(penguins.describe()) # Średnie i std dają obraz skali danych (np. masa ok. 4.2kg).
display(penguins.describe(include='all')) # Tekstowe dodają: unique, top (najczęstszy), freq.
print(f"Mediana masy ciała: {penguins['body_mass_g'].median()}")

# Zadanie 3c: Rozkłady wartości
print("\n--- Zadanie 3c ---")
print("Gatunki:")
print(penguins['species'].value_counts())
print("\nWyspy:")
print(penguins['island'].value_counts())
print("\nPłeć:")
print(penguins['sex'].value_counts()) # Rozkład jest niemal idealnie zbalansowany (168 vs 165).
print("\nProcentowy udział gatunku Adelie:")
print(penguins['species'].value_counts(normalize=True)['Adelie'] * 100)

# Zadanie 3d: Brakujące dane
print("\n--- Zadanie 3d ---")
brakujace = penguins.isna().sum()
print("Braki w kolumnach:")
print(brakujace)

procent_brakow_sex = (penguins['sex'].isna().sum() / len(penguins)) * 100
print(f"\nProcent brakujących danych w 'sex': {procent_brakow_sex:.2f}%")
```

```python
# Zadanie 4
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
# Dodaj kolumnę z procentem napiwku
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(1)


# 1. Oblicz średni rachunek (total_bill)
sredni_rachunek = tips['total_bill'].mean()
print(f"Średni rachunek: {sredni_rachunek:.2f}")

# 2. Oblicz średni napiwek w procentach (tip_pct)
# Najpierw tworzymy nową kolumnę, potem liczymy średnią
tips['tip_pct'] = (tips['tip'] / tips['total_bill']) * 100
print(f"Średni napiwek (%): {tips['tip_pct'].mean():.2f}%")

# 3. Wyświetl który dzień ma najwięcej rachunków
print("\nLiczba rachunków wg dni:")
print(tips['day'].value_counts())

# 4. Policz ile rachunków jest od palaczy
print("\nLiczba palaczy:")
print(tips['smoker'].value_counts())

# 5. Wyświetl describe() tylko dla kolumny 'tip'
print("\nStatystyki dla kolumny 'tip':")
display(tips['tip'].describe())
print(f"Mediana napiwku: {tips['tip'].median()}")

# 6. Jaki procent rachunków pochodzi z obiadu (Dinner)?
procent_time = tips['time'].value_counts(normalize=True) * 100
print(f"\nProcent rachunków 'Dinner': {procent_time['Dinner']:.2f}%")

# 7. Ile unikalnych rozmiarów grup jest w danych?
unikalne_grupy = tips['size'].nunique()
print(f"\nLiczba unikalnych rozmiarów grup: {unikalne_grupy}")
```
